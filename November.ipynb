Write a function sum_of_even_numbers that takes a list of integers as input and returns the sum of all even numbers in the list.

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total = total + num
    return total

sum_of_even_numbers([1,2,3,4,5])
```

<h2>2024.11.05</h2>
Find even numbers

```python
def find_even(numbers):
    even_numbers = []
    for num in numbers:
        if num % 2 == 0:
            even_numbers.append(num)
    return even_numbers
            
find_even([1,2,3,4,5,6,7])
```

How to compare two NumPy arrays?

```python
import numpy as np

an_array = np.array([[1, 2], [3, 4]])
another_array = np.array([[1, 2], [3, 4]])

comparison = an_array == another_array
equal_arrays = comparison.all()

print(equal_arrays)

#Sol2
import numpy as np

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])


# Comparing the arrays
if np.array_equal(arr1, arr2):
	print("Equal")
else:
	print("Not Equal")

```
