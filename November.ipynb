Write a function sum_of_even_numbers that takes a list of integers as input and returns the sum of all even numbers in the list.

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total = total + num
    return total

sum_of_even_numbers([1,2,3,4,5])
```
